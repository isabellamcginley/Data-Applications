```python
# prompt: compare discounts vs sales in a line graph with two different lines representing each one

import matplotlib.pyplot as plt

# Define the data
discounts = [0.1, 0.15, 0.2, 0.25, 0.3]
sales = [100, 120, 150, 180, 200]

# Create the plot
plt.plot(discounts, sales, label="Sales")

# Create the second plot with a different color
plt.plot(discounts, [sale * 0.9 for sale in sales], label="Discounts", color="red")

# Add labels and title
plt.xlabel("Discount")
plt.ylabel("Sales")
plt.title("Discounts vs Sales")

# Add legend
plt.legend()

# Show the plot
plt.show()
```

```python
# prompt: create a scatter plot for discount on top of a line plot for average sales per day with xaxis being dates on top of line

import matplotlib.pyplot as plt
# Create some data
dates = ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05']
average_sales_per_day = [100, 120, 150, 180, 200]
discounts = [0.1, 0.15, 0.2, 0.25, 0.3]

# Create the line plot
plt.plot(dates, average_sales_per_day, label="Average Sales per Day")

# Create the scatter plot for discounts
plt.scatter(dates, discounts, label="Discounts", color="red")

# Add labels and title
plt.xlabel("Date")
plt.ylabel("Value")
plt.title("Discounts vs Average Sales per Day")

# Add legend
plt.legend()

# Show the plot
plt.show()
```
